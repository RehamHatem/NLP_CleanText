Extract HTML from URL

```python
import urllib.request

url = "https://huggingface.co/nlpconnect/vit-gpt2-image-captioning"

# First open, then read the the HTML
html = urllib.request.urlopen(url).read()

# Print the HTML code
print(html)

# prints the datatype of the output.
print(type(html))
```

Extract TEXT from HTML page

```python
from bs4 import BeautifulSoup

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
# Extract text from the parsed HTML
text = soup.get_text()
print(text)
```

Processing on Data

```python
from nltk.corpus import stopwords

# normalization
def normalize_text(text):
    normalized_text = ''
    for char in text:
        if 'A' <= char <= 'Z':
            normalized_text += chr(ord(char) + 32)  #ord=>Unicode code , Adding 32=>convert to lowercase
        else:
            normalized_text += char
    return normalized_text

# tokenization
def tokenize_text(text):
    words = []
    current_word = ''
    for char in text:
        if char.isalpha():
            current_word += char
        elif current_word:
            words.append(current_word)
            current_word = ''
    if current_word:
        words.append(current_word)
    return words

#  lemmatization
def lemmatize_words(words):
    lemmatized_words = []
    for word in words:
        if word.endswith('s'):
            lemmatized_words.append(word[:-1])
        else:
            lemmatized_words.append(word)
    return lemmatized_words

#  removing stop words

# def remove_stop_words(words):
#     stop_words = set(["a", "an", "the", "and", "or", "but", "if", "then", "else", "for", "of", "with", "on", "at", "by", "about", "to","alfasign"])
#     filtered_words = []
#     for word in words:
#         if word not in stop_words:
#             filtered_words.append(word)
#     return filtered_words

def remove_stop_words(words):
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in lemmatized_words if word not in stop_words]
    return filtered_words


def remove_three_character_words(words):
    return [word for word in words if len(word) > 3]

def print_three_character_words(words):
    three_char_words = [word for word in words if len(word) == 3]
    for word in three_char_words:
        print(word)
    print("====> number of three char words :")
    print(len(three_char_words))



normalized_text = normalize_text(text)
tokenized_text = tokenize_text(normalized_text)
lemmatized_words = lemmatize_words(tokenized_text)
filtered_words = remove_stop_words(lemmatized_words)
remove_three_caracters_words = remove_three_character_words(filtered_words)


print("===> (before removing three chars words: )")
before_removing_three_chars_words = ' '.join(filtered_words)
print(before_removing_three_chars_words)

print("===> (after removing three chars words:)")
processed_text = ' '.join(remove_three_caracters_words)
print(processed_text)

print("====> (Words with three characters: )")
print_three_chars_words = print_three_character_words(filtered_words)
```

Unique Words

```python
def get_unique_words(text):
    # Initialize an empty set to store unique words
    unique_words = set()
    # Split text into words
    words = text.split()
    # Iterate over words and add them to the set
    for word in words:
        unique_words.add(word)
    return unique_words


print("===> (unique words before removing three chars words: )")
unique_words = get_unique_words(before_removing_three_chars_words)
print(unique_words)
print("Number of all unique words:", len(unique_words))

print("===> (unique words after removing three chars words: )")
unique_words = get_unique_words(processed_text)
print(unique_words)
print("Number of unique words:", len(unique_words))


```

```python
import re
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

def extract_and_process_url(url):
    # Extract HTML from the URL
    response = requests.get(url)
    html_content = response.text


    soup = BeautifulSoup(html_content, 'html.parser')
    text = ' '.join(soup.stripped_strings)


    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)

    normalized_text = cleaned_text.lower()

    tokens = word_tokenize(normalized_text)

    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in lemmatized_tokens if word not in stop_words]

    unique_words = set(word for word in filtered_tokens if len(word) < 3)

    return unique_words
my=extract_and_process_url("https://huggingface.co/nlpconnect/vit-gpt2-image-captioning")
print(my)
```
